```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
```

```python
# digits_wine_load_breast_cancer 분류 함수
def digits_wine_load_breast_cancer(data,label):
  # train, test 데이터 분리
  X_train, X_test, y_train, y_test = train_test_split(data,
                                                    label,
                                                    test_size=0.2,
                                                    random_state=15)
  # 모델 학습 및 예측
  #Desicion Tree 사용
  decision_tree = DecisionTreeClassifier(random_state=15)
  decision_tree.fit(X_train, y_train)
  y_pred = decision_tree.predict(X_test)
  print(classification_report(y_test, y_pred))

  #SVM 사용
  svm_model = DecisionTreeClassifier(random_state=20)
  svm_model.fit(X_train, y_train)
  y_pred = svm_model.predict(X_test)
  print(classification_report(y_test, y_pred))

  #랜덤 포레스트
  random_forest = RandomForestClassifier(random_state=30)
  random_forest.fit(X_train, y_train)
  y_pred = random_forest.predict(X_test)
  print(classification_report(y_test, y_pred))
    
  #SGD Classifier 사용해 보기
  sgd_model = RandomForestClassifier(random_state=40)
  sgd_model.fit(X_train, y_train)
  y_pred = sgd_model.predict(X_test)
  print(classification_report(y_test, y_pred))
    
  #Logistic Regression 사용
  logistic_model = RandomForestClassifier(random_state=50)
  logistic_model.fit(X_train, y_train)
  y_pred = logistic_model.predict(X_test)
  print(classification_report(y_test, y_pred))

  return print(confusion_matrix(y_test, y_pred))
```

```python
# digtis 분류
digits = load_digits()
digits_data = digits.data
digits_label = digits.target
digits_wine_load_breast_cancer(digits_data,digits_label)

print(digits.target_names)

# wine 분류
wine = load_wine()
wine_data = wine.data
wine_label = wine.target
digits_wine_load_breast_cancer(wine_data,wine_label)

print(wine.target_names)


# load_breast_cancer 분류
load_breast_cancer = load_breast_cancer()
load_breast_cancer_data = load_breast_cancer.data
load_breast_cancer_label = load_breast_cancer.target

digits_wine_load_breast_cancer(load_breast_cancer_data,load_breast_cancer_label)

print(load_breast_cancer.target_names)
```

#### **학습된 모델들의 테스트데이터 예측 결과를 어떻게 해석해야 할까요?**

예측 결과 정확도는 정답의 분포에 따라 모델의 성능을 잘 평가하지 못하는 척도가
될 수 있으므로 정확도 외의 precision, recall 등 다양한 평가 척도를 사용하여 해석해야 한다.

#### **모델의 성능을 평가하는 지표로는 무엇이 좋을까요?**  

precision과 recall을 가지고 이 분류는 어느 지표가 더 중요한지 판단하여
하나의 지표를 고르는 것이 좋은 것 같다.
1번째 숫자에서는 비밀번호 등에 사용된다면 오답인것을 정답이란 한것이 더 큰 문제이기 때문에 pecision이 더 중요하고 생각한다.
2번째 와인에서는 올바르지 않은 와인을 정답 와인으로 분류하면 안되기 때문에
precision이 더 중요하다고 생각한다.
3번째 유방암은 암을 진단하는 경우 실제 환자를 한 명이라도 놓치면 안 된다. 즉 양성을 음성으로 판단하면 안 되기 때문에 Recall이 더 중요하다고 생각한다.
